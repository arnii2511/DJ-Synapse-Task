### The previous commands were just to make sure we have a workable dataframe now, we'll do some manipulations on the data.

Find out the top 5 most voted candidates in Telangana

```python
df_telan=df[df['State']=='Telangana'] #sorting acc to state then in desc order
df_telan_sort=df_telan.sort_values(by='Total Votes',ascending=False)
top_five = df_telan_sort.head(5)
print(top_five)
```

Find out the total number of votes received by each state

```python
#using groupby
total_votes=df.groupby('State')['Total Votes'].sum().reset_index()
total_votes.columns=['State','Total Votes']
print(total_votes)
```

Show these values as a percentage of the total number of votes received in the election.

```python
total_votecount=total_votes['Total Votes'].sum()
total_votes['Percent Share']=(total_votes['Total Votes']/total_votecount)*100
print(total_votes)
```

Now, display the top 5 parties and find out the number of votes received by them

```python
party_group=df.groupby('Party')['Total Votes'].sum().reset_index()
party_group.columns=['Party','Total Votes']
party_group_sort=party_group.sort_values(by='Total Votes',ascending=False)
print(party_group_sort.head(5))
```

From the candidates who lost, list 10 candidates who got the highest percent votes.

```python
candi_lost=df[df['Result']=='Lost']
lost_group=candi_lost.sort_values(by='% of Votes',ascending=False)
print(lost_group.head(10)['Candidate'])
```

### These techniques help to understand the data that we have but we like something a bit more visual right? That's where matplotlib comes to our aid.

Make a piechart to show the votes recieved by the top 5 candidates in the elections.

```python
top_candi=df.sort_values(by='Total Votes',ascending=False)
top_five_candi=top_candi.head(5)
labels=top_five_candi['Candidate']
sizes=top_five_candi['Total Votes']
fig,ax=plt.subplots()
ax.pie(sizes,labels=labels,autopct="%1.1f%%")
ax.axis('equal')
plt.title("Top 5 Candidates")
plt.show()
```

Make a bar graph to see the postal votes received by the parties in Karnataka

```python
df_karna=df[df['State']=='Karnataka']
plt.figure(figsize=(50,50))
plt.bar(df_karna['Party'],df_karna['Postal Votes'])
plt.xlabel('Party')
plt.ylabel('Postal Votes')
plt.title("Postal Votes received by the parties in Karnataka")
plt.xticks(rotation=45,ha='right')
plt.tight_layout()
plt.show()

### Congratulations on making it to the end. EDA is a very vital toolkit to tackle Machine Learning problems. Hopefully this task helped you understand concepts of data analysis and visualisation. See you at the interviews :)
```
